# Training performance with different board sizes

In this notebook I will study how the board size inpacts on training performance.



```python
import pandas as pd
import os.path

# Configuration
stats_directory=os.path.join("..", "stats")

# Boards
boards = ["4x4", "5x5", "6x6"]

for board_size in boards:
    stat_file=os.path.join(stats_directory, f"stats-{board_size}.csv")
    
    stat_df = pd.read_csv(stat_file)
    
    print(f"--- > Board size {board_size}")
    print(stat_df)
```
